```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Weather Data.csv")
```

```python
df.head(3)
```

```python
df=df.drop("Date/Time", axis=1)
```

```python
df=df.drop("Weather", axis=1)
```

```python
df.head(3)
```

```python
plt.figure(figsize=(12,8))
plt.title("Temperature")
plt.xlabel("Harorat (selsiyda)")
sns.histplot(df["Temp_C"], kde=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Temperature")
plt.xlabel("Harorat (selsiyda)")
rebust_scaling = (df-df.median())/(df.quantile(0.75)-df.quantile(0.25))
sns.histplot(rebust_scaling["Temp_C"], kde=True, color="RED")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Temperature")
plt.xlabel("Harorat (selsiyda)")
max_abs_scaling = df/df.abs().max()
sns.histplot(max_abs_scaling["Temp_C"], kde=True, color="k")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Temperature")
plt.xlabel("Harorat (selsiyda)")
mean_normalization = (df-df.mean())/(df.max()-df.min())
sns.histplot(mean_normalization["Temp_C"], kde=True, color="y")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Temperature")
plt.xlabel("Harorat (selsiyda)")
min_max_scaling = (df-df.min())/(df.max()-df.min())
sns.histplot(min_max_scaling["Temp_C"], kde=True, color="blue")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Temperature")
plt.xlabel("Harorat (selsiyda)")
standardisation = (df-df.mean())/df.std()
sns.histplot(standardisation["Temp_C"], kde=True, color="orange")
plt.show()
```
